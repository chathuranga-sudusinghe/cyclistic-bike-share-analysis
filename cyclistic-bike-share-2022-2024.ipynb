```python
path_2023 = "/kaggle/input/cyclistic-bikeshare-2022-2024/clean_cyclistic_2023/clean_cyclistic_2023.csv"
df_2023 = pd.read_csv(path_2023)

df_2023.shape
```

```python
path_2024 = "/kaggle/input/cyclistic-bikeshare-2022-2024/clean_cyclistic_2024/clean_cyclistic_2024.csv"
df_2024 = pd.read_csv(path_2024)

df_2024.shape
```

```python
df_all = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)
df_all.shape
```

```python
# Fill missing station names and IDs safely
df_all['start_station_name'] = df_all['start_station_name'].fillna("Unknown")
df_all['end_station_name']   = df_all['end_station_name'].fillna("Unknown")
df_all['start_station_id']   = df_all['start_station_id'].fillna("Unknown")
df_all['end_station_id']     = df_all['end_station_id'].fillna("Unknown")
```

```python
df_all[['start_station_name','end_station_name','start_station_id','end_station_id']].isnull().sum()
```

```python
# Convert started_at and ended_at to datetime
df_all['started_at'] = pd.to_datetime(df_all['started_at'], errors='coerce')
df_all['ended_at'] = pd.to_datetime(df_all['ended_at'], errors='coerce')
```

```python
output_path = "/kaggle/working/cyclistic_2022_2024_MASTER_CLEAN.csv"

df_all.to_csv(output_path, index=False)

output_path
```
